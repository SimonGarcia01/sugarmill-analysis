```python
dfHS = dfHS[dfHS.columns[dfHS.isna().sum() <= 16497]]
```

```python
dfHS.shape
```

```python
dfHS_no_newmetrics= dfHS.drop(dfHS.columns[dfHS.isna().sum() > 16496], axis=1)
dfHS_no_newmetrics.shape
```

```python
dfHS_newmetrics = dfHS[dfHS['Evaporacion 0-3'].isna() == False]
dfHS_newmetrics.shape
```

## EDA New Metrics DF

```python
dfHS_newmetrics.info()
```

```python
dfHS_newmetrics.describe()
```

Dataframe separation in numerical and no-numerical data types

```python
_numerical = dfHS_newmetrics.select_dtypes(include=['number'])

_no_numerical = dfHS_newmetrics.select_dtypes(exclude=['number'])
```

Obtaining missing values

numerical

```python
missing_values = _numerical.isna().sum()

columns_with_missing_values = missing_values[missing_values > 0]

print(columns_with_missing_values)
```

no-numerical

```python
missing_values = _no_numerical.isna().sum()

columns_with_missing_values = missing_values[missing_values > 0]

print(columns_with_missing_values)
```

Replace missing values

numerical with median and mean method according to Skewness and Kurtosis

Calculate skewness and kurtosis

```python
skewness = _numerical.skew()
kurtosis = _numerical.kurtosis()

print("Skewness for column:")
print(skewness)
print("\kurtosis for column:")
print(kurtosis)
```

```python
for col in _numerical.select_dtypes(include=['float64', 'int64']).columns:
    if abs(skewness[col]) > 1 or kurtosis[col] > 3:
        _numerical[col] = _numerical[col].fillna(_numerical[col].median())  # Replace for median
    else:
        _numerical[col] = _numerical[col].fillna(_numerical[col].mean())  # Replace for mean
```

no-numerical according to mode

```python
for col in _no_numerical.columns:
    _no_numerical[col] = _no_numerical[col].fillna(_no_numerical[col].mode()[0])  # Reemplaza for mode
```

Review missing values

```python
missing_values = _no_numerical.isna().sum()

columns_with_missing_values = missing_values[missing_values > 0]

print(columns_with_missing_values)
```

```python
missing_values = _numerical.isna().sum()

columns_with_missing_values = missing_values[missing_values > 0]

print(columns_with_missing_values)
```

```python
from sklearn.decomposition import PCA

# Reducir a 2 dimensiones con PCA para visualización
pca = PCA(n_components=2)
_numerical_reduced = pca.fit_transform(_numerical)

# Graficar las dos primeras componentes principales
plt.figure(figsize=(8, 6))
plt.scatter(_numerical_reduced[:, 0], _numerical_reduced[:, 1])
plt.title('PCA de las variables numéricas')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)  # Supongamos que queremos 3 clusters
_newmetrics_numerical_reduced_clusters = kmeans.fit_predict(_numerical_reduced)

plt.scatter(_numerical_reduced[:, 0], _numerical_reduced[:, 1], c=_newmetrics_numerical_reduced_clusters)
plt.title('PCA y Clusterización K-Means')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()
```

```python
plt.hist(_numerical_reduced[:, 0], bins=30, alpha=0.5, label="Componente principal 1")
plt.hist(_numerical_reduced[:, 1], bins=30, alpha=0.5, label="Componente principal 2")
plt.legend(loc='upper right')
plt.title('Distribución de las componentes principales')
plt.show()
```
